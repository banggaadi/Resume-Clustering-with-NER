```python
import pandas as pd

data = pd.read_csv('./Dataset/dataset_skill.csv', delimiter=';')

# Fill the blanks
data['Sentence #'].fillna(method='ffill', inplace=True)
```

## Convert to spacy format

```python
sentences = []
current_sentence = []
current_sentence_num = None
for _, row in data.iterrows():
    if row['Sentence #'] != current_sentence_num:
        if current_sentence:
            sentences.append(current_sentence)
        current_sentence = []
        current_sentence_num = row['Sentence #']
    current_sentence.append((row['Word'], row['Tag']))

if current_sentence:
    sentences.append(current_sentence)

# Convert to spaCy format
def convert_to_spacy_format(sentences):
    training_data = []
    for sentence in sentences:
        words = [word for word, tag in sentence]
        entities = []
        start = 0
        for word, tag in sentence:
            end = start + len(word)
            if tag != 'O':
                entities.append((start, end, tag))
            start = end + 1
        training_data.append({"tokens": words, "entities": entities})
    return training_data

data = convert_to_spacy_format(sentences)
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
```

```python
import json

with open('./Dataset/train_data.json', 'w', encoding='utf-8') as f:
    json.dump(train_data, f, ensure_ascii=False, indent=4)

with open('./Dataset/test_data.json', 'w', encoding='utf-8') as f:
    json.dump(test_data, f, ensure_ascii=False, indent=4)
```
